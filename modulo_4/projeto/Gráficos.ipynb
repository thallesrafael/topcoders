# Configurações e importações

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings 
```

```python
pd.set_option("display.max_columns", 100)
plt.rcParams["figure.figsize"] = (15, 8)
sns.set(rc={"figure.figsize": (15, 8), 'axes.titlesize': 16, 'xtick.labelsize': 13, 'ytick.labelsize': 13})
sns.set_style('darkgrid')
warnings.filterwarnings('ignore')
```

```python
df_customer = pd.read_csv('Top-Coders-Project4/data/marketing_campaign.csv', sep='\t')
```

```python
df_customer.drop(['Z_Revenue', 'Z_CostContact'], axis=1)
```

```python
df_customer.head(10)
```

```python
df_customer.info()
```

```python
correcao_nomes = {'YOLO':'Single',
            'Alone':'Single',
            'Absurd' :'Single',}

df_customer = df_customer.replace(correcao_nomes)
```

## Adicionando novas colunas

```python
df_customer['MntTotal'] = df_customer['MntWines'] + df_customer['MntFruits'] + df_customer['MntMeatProducts'] + df_customer['MntFishProducts'] + df_customer['MntSweetProducts'] + df_customer['MntGoldProds']
df_customer['accepted_offer'] = (df_customer.AcceptedCmp1 + df_customer.AcceptedCmp2 + df_customer.AcceptedCmp3 + df_customer.AcceptedCmp4 + df_customer.AcceptedCmp5).apply(lambda x: 0 if x == 0 else 1)
df_customer['have_sons'] = (df_customer['Kidhome'] + df_customer['Teenhome']).apply(lambda x: 0 if x == 0 else 1)
df_customer['Age'] = 2015 - df_customer['Year_Birth']
```

## Criando classificações

### Idade

```python
max_age = df_customer['Age'].max()
min_age = df_customer['Age'].min()
labels = ['Adults', 'Middle Age', 'Elder']
classes = [min_age, 40, 60, max_age]

df_customer['age_group'] = pd.cut(x = df_customer['Age'], bins = classes, labels = labels, include_lowest = True)
```

```python
print(f'Cliente mais velho possui {max_age} anos.')
print(f'Cliente mais novo possui {min_age} anos.')
```

### Renda

```python
labels = ['Low Income', 'Middle Income', 'High Income']
classes = [0, 30000, 60000, 1000000]

df_customer['Income_group'] = pd.cut(x = df_customer['Income'], bins = classes, labels = labels, include_lowest = True)
```

## Realizando agrupamentos para estudo

```python
marital_status_grouping = df_customer.groupby(by = 'Marital_Status')
sons_or_not_grouping = df_customer.groupby(by = 'have_sons')
age_grouping = df_customer.groupby(by = 'age_group')
Income_grouping = df_customer.groupby(by = 'Income_group')
Education_grouping = df_customer.groupby(by = 'Education')
```

## Local de compra de preferência por grupo

```python
local_per_age_group = age_grouping.mean()[['NumWebPurchases', 'NumDealsPurchases',
                                                       'NumCatalogPurchases', 'NumStorePurchases']].T
for i in local_per_age_group.columns:
    local_per_age_group[i] = local_per_age_group[i]/local_per_age_group[i].sum()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12,4), sharey=True)

explode = [0, 0, 0, 0.1] 

axs[0].pie(local_per_age_group['Adults'], autopct='%.1f%%', explode = explode, shadow = True)
axs[0].set(title='Jovens Adultos')

axs[1].pie(local_per_age_group['Middle Age'], autopct='%.1f%%', explode = explode, shadow = True)
axs[1].set(title='Adultos')

axs[2].pie(local_per_age_group['Elder'], autopct='%.1f%%', explode = explode, shadow = True)
axs[2].set(title='Idosos')

plt.legend(local_per_age_group.index, ncol = 1, bbox_to_anchor = (1, 0.7))
```
