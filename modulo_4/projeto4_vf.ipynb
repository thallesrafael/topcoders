#### Sobre o conjunto de dados


<font size="2">A Análise da Personalidade do Cliente é uma análise detalhada dos clientes ideais de uma empresa. Ajuda uma empresa a entender melhor seus clientes e facilita a modificação de produtos de acordo com as necessidades, comportamentos e preocupações específicas de diferentes tipos de clientes.  
A análise da personalidade do cliente ajuda uma empresa a modificar seu produto com base em seus clientes-alvo de diferentes tipos de segmentos de clientes. Por exemplo, em vez de gastar dinheiro para comercializar um novo produto para cada cliente no banco de dados da empresa, uma empresa pode analisar qual segmento de cliente tem mais probabilidade de comprar o produto e, em seguida, comercializar o produto apenas nesse segmento específico.</font>

##### Conteúdo

<font size="2">

*Pessoas*:  
* ID:  identificador exclusivo do cliente  
* Year_Birth: ano de nascimento do cliente  
* Educação: nível de escolaridade do cliente  
* Marital_Status: estado civil do cliente  
* Renda: renda familiar anual do cliente  
* Kidhome: número de crianças na casa do cliente  
* Teenhome: número de adolescentes na casa do cliente  
* Dt_Customer: Data de inscrição do cliente na empresa  
* Recência: número de dias desde a última compra do cliente  
* Reclame: 1 se o cliente reclamou nos últimos 2 anos, 0 caso contrário  

*Produtos*  
* MntWines: Valor gasto em vinhos nos últimos 2 anos  
* MntFruits: valor gasto em frutas nos últimos 2 anos  
* MntMeatProducts: valor gasto em carnes nos últimos 2 anos  
* MntFishProducts: Valor gasto em pescado nos últimos 2 anos  
* MntSweetProducts: Valor gasto em doces nos últimos 2 anos  
* MntGoldProds: Valor gasto em ouro nos últimos 2 anos  

*Promoção*  
* NumDealsPurchases:* Número de compras feitas com desconto
* AcceptedCmp1: 1 se o cliente aceitou a oferta na 1ª campanha, 0 caso contrário  
* AcceptedCmp2: 1 se o cliente aceitou a oferta na 2ª campanha, 0 caso contrário  
* AcceptedCmp3: 1 se o cliente aceitou a oferta na 3ª campanha, 0 caso contrário  
* AcceptedCmp4: 1 se o cliente aceitou a oferta na 4ª campanha, 0 caso contrário  
* AcceptedCmp5: 1 se o cliente aceitou a oferta na 5ª campanha, 0 caso contrário  
* Resposta: 1 se o cliente aceitou a oferta na última campanha, 0 caso contrário  

*Lugar*  
* NumWebPurchases: Número de compras feitas pelo site da empresa  
* NumCatalogPurchases: Número de compras feitas usando um catálogo  
* NumStorePurchases: Número de compras feitas diretamente nas lojas  
* NumWebVisitsMonth: Número de visitas ao site da empresa no último mês  

*Link da fonte do conjunto de dados*: https://www.kaggle.com/datasets/imakash3011/customer-personality-analysis/discussion</font>

## Limpeza e Ajustes do DataFrame

```python
# Retirando os valores nulos da coluna Income

df_customer.dropna(inplace=True)

# Limpeza de algumas colunas

df_customer.drop(['Z_Revenue', 'Z_CostContact'], axis=1)

# Correção dos Nomes
correcao_nomes = {'YOLO': 'Single',
                'Alone': 'Single',
                'Absurd': 'Single'}

df_customer = df_customer.replace(correcao_nomes)
```

## Criação de Novas Colunas

```python
# Criando uma coluna com o total de filhos de cada cliente
df_customer['Total_Children'] = df_customer['Kidhome'] + df_customer['Teenhome']

# Criando uma coluna para indicar se um cliente tem ou não filhos
df_customer['Have_sons'] = df_customer['Total_Children'].apply(lambda x: 'Não' if x == 0 else 'Sim')

# Criando uma coluna com o valor total gasto de cada cliente
df_customer['MntTotal'] = df_customer['MntWines'] + df_customer['MntFruits'] + df_customer['MntMeatProducts'] + df_customer['MntFishProducts'] + df_customer['MntSweetProducts'] + df_customer['MntGoldProds']

# Criando uma coluna para indicar se um cliente aceitou ou não uma das ofertas
df_customer['Accepted_offer'] = df_customer['AcceptedCmp1'] + df_customer['AcceptedCmp2'] + df_customer['AcceptedCmp3'] + df_customer['AcceptedCmp4'] + df_customer['AcceptedCmp5']
df_customer['Accepted_offer'] = df_customer['Accepted_offer'].apply(lambda x: 0 if x == 0 else 1)

# Passando a coluna 'Dt_Customer' para o formato datatime
df_customer['Dt_Customer'] = pd.to_datetime(df_customer['Dt_Customer'], format = '%d-%m-%Y')

# Criando a coluna para representar a idade de cada cliente (foi considerado o ano posterior ao do último cliente cadastrado)
df_customer['Age'] = 2015 - df_customer['Year_Birth']
```

## Removendo Outliers

```python
def outliers(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_boundary = Q1 - 1.5 * IQR
    upper_boundary = Q3 + 1.5 * IQR
    df_customer = dataframe.loc[(dataframe[column] >= lower_boundary) & (dataframe[column] <= upper_boundary)]
    return df_customer

# Retirando outliers da coluna referente ao ano de nascimento
df_customer = outliers(df_customer, 'Year_Birth')

# Optamos por retirar apenas o cliente com renda incompatível
income_outlier = df_customer[df_customer['Income']==666666.0].index
df_customer.drop(income_outlier, inplace=True)
```

## Qual a relação entre o consumo de cada produto e escolaridade?

```python
# gráfico para entender que tipo de produto é interessante para cada grau de escolaridade dos clientes
columns_Mnt = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df_educ_prod_mean = df_customer.pivot_table(index='Education', values=columns_Mnt, aggfunc='mean')
df_educ_prod_mean.rename(mapper = {'MntWines': 'Vinhos', 'MntFruits': 'Frutas', 'MntMeatProducts': 'Carnes', 'MntFishProducts': 'Peixes', 'MntSweetProducts': 'Doces', 'MntGoldProds':'Ouro'}, inplace=True, axis = 1)
df_educ_prod_mean = df_educ_prod_mean.reindex(['Basic', '2n Cycle', 'Graduation', 'Master', 'PhD']).rename({'Graduation': 'Graduado', 'PhD': 'Doutor', 'Master': 'Mestre', '2n Cycle': '2° Ciclo', 'Basic': 'Básico'})
fig = px.bar(df_educ_prod_mean, barmode='group', color_discrete_sequence=px.colors.qualitative.Vivid)

fig.update_layout(
    xaxis_title=None,
    yaxis_title={
        'text': "Média do valor gasto",
        'font_size': 13
    },
    legend_title={
        'text': "Produtos:",
        'font_size': 13
    },
    template='none',
    title={
        'text': 'Média do valor gasto por produtos e escolaridade',
        'y':0.95,
        'x':0.25,
        'font_color': 'black',
        'font_size': 18
    }
)
```

##### Observações: 
É possível notar que o produto mais comprado é o vinho seguindo pela carne, nota-se também que a compra média de vinho tem um crescimento de acordo com o aumento do grau escolar.
